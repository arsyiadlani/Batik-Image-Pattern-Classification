### Google Images Scraper

Pastikan library sudah terinstall terlebih dahulu dengan menjalankan line berikut pada cmd: \
**pip install git+https://github.com/Joeclinton1/google-images-download.git**

Masukkan keyword gambar yang ingin dicari pada arguments -> "keywords" dan masukkan jumlah pencarian pada arguments -> "limit". Nantinya akan muncul folder 'downloads' pada lokasi path yang sama dengan lokasi notebook ini dijalankan. Dalam folder 'downloads' tersebut akan ada folder-folder untuk hasil pencarian setiap keyword yang dimasukkan. (Limit maksimal hasil pencarian adalah 100 gambar)

```python
from google_images_download import google_images_download

# Instansiasi kelas
response = google_images_download.googleimagesdownload()
arguments = {"keywords":"batik truntum",
             "limit":100,"print_urls":False}
paths = response.download(arguments)

# Print judul (path) gambar yang didownload
print(paths)
```

### Bing Images Scraper

Pastikan library sudah terinstall terlebih dahulu dengan menjalankan line berikut pada cmd: \
**pip install bing-image-downloader**

Hasil pencarian gambar akan terdapat pada folder target 'dataset'. Keunggulan dari Images Scraper untuk search engine Bing ini adalah tidak ada batasan limit maksimal gambar hasil pencarian.

```python
from bing_image_downloader import downloader

downloader.download("motif kawung", limit=200,  output_dir='dataset', 
                    adult_filter_off=True, force_replace=False, timeout=60)
downloader.download("batik kawung", limit=200,  output_dir='dataset', 
                    adult_filter_off=True, force_replace=False, timeout=60)
downloader.download("kawung", limit=200,  output_dir='dataset', 
                    adult_filter_off=True, force_replace=False, timeout=60)
```

### Renaming Downloaded Images

Mengubah nama file-file gambar hasil scraping sehingga lebih rapi dan beraturan (i.e.: Kawung 1, Kawung 2, ...)

```python
import os

path = 'C:/Users/dell/Documents/SDR/downloads/motif truntum'  # Path lokasi folder gambar hasil pencarian
images = os.listdir(path)
print(f'Total gambar: {len(images)} gambar')

os.chdir(path)

# Penamaan ulang gambar menjadi Kawung 1, Kawung 2, ..., Kawung X; dengan X adalah total gambar dalam folder
i = 1
for file in images:
    os.rename(file, 'Truntum_'+ str(i)+'.jpg')
    i = i+1
```
